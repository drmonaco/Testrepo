```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm


url = "https://raw.githubusercontent.com/ds4ph-bme/homework-6-drmonaco/master/shhs1.txt?token=AITD5B223XIYMQ3R5MXGNQS6RNNYS"
df = pd.read_csv(url, error_bad_lines=False,delimiter = '\t')
df2 = df[['rdi4p','bmi_s1']]

log_rd14p = np.log(df2.rdi4p+1)

df1 = pd.DataFrame({'x': log_rd14p})
df1['y'] = df2.bmi_s1
df1 = df1.dropna()
y = df1.x.values.reshape(-1,1)
x = df1.y.values.reshape(-1,1)



model = LinearRegression() 
results = model.fit(x, y) #training the algorithm
print("The slope of BMI vs log(rdi4p +1 is")
print(results.coef_)
print("The intercept of BMI vs log(rdi4p +1 is")
print(results.intercept_)

plt.scatter(x, y )
Y_pred = model.predict(x)
plt.plot(x, Y_pred, color='red')

plt.show()

log_rd14p_test = 30*results.coef_+results.intercept_
print("If BMI is 30 then log(rdi4p +1 is")

print(log_rd14p_test)
print("If BMI is 30 then rdi4p")

print(np.exp(log_rd14p_test)-1)
```
